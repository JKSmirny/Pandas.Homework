```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Optional: give the displayed data cleaner formatting

```python
#Calculate number of schools in the disrict 
school_data_complete["school_name"].nunique()
total_schools = school_data_complete["school_name"].nunique()

#Calculate number of students in the district
total_students = len(school_data_complete["student_name"])

#Calculate total district's budget
school_data_complete["budget"].sum()
total_budget = school_data_complete["budget"].sum()

#Calculate average math score
school_data_complete["math_score"].mean()
average_math_score = school_data_complete["math_score"].mean()

#Calculate average reading score
school_data_complete["reading_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()

#Create a table w/number of students with passing math scores
#pass_math_score = school_data_complete.loc[school_data_complete['math_score'] >= 70]
#pass_math_score

#Calculate number of students with passing math scores using loc function
pass_math_score = school_data_complete.loc[school_data_complete['math_score'] >= 70]

#Calculate the percentage of students who passed math
percent_math = len(pass_math_score)/total_students*100

#Create a table w/number of students who passed reading
#pass_reading_score = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
#pass_reading_score

#Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading_score = school_data_complete.loc[school_data_complete['reading_score'] >= 70]

#Calculate the percentage of students who passed reading
percent_reading = len(pass_reading_score)/total_students*100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_rate = (percent_math + percent_reading)/2


```
